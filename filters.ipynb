```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
cv2.__version__
```

```python
!ls
img = cv2.imread("shoutao.JPG")
img.shape
```

```python
img_resized = cv2.resize(img, (216, 288))
img = img_resized
cv2_imshow(img_resized)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

```python
img_blur = cv2.GaussianBlur(img_gray, (15, 15), sigmaX = 0)
cv2_imshow(img_blur)
img_median_blur = cv2.medianBlur(img_gray, ksize=9)
cv2_imshow(img_median_blur)
```

```python
img_canny = cv2.Canny(img_blur, 70, 70)
cv2_imshow(img_canny)
```

```python
kernel = np.ones((3,3), np.uint8)
img_dil = cv2.dilate(img_canny, kernel, iterations = 1)
cv2_imshow(img_dil)
```
